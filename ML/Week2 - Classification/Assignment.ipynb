```python
import graphlab
```

```python
products = graphlab.SFrame('amazon_baby.gl/')
```

```python
products['word_count'] = graphlab.text_analytics.count_words(products['review'])
```

```python
products.head()
```

```python
def awesome_fun(d):
    if 'awesome' in d:
        count = d['awesome']
    else:
        count = 0
    return count
```

```python
products['awesome'] = products['word_count'].apply(awesome_fun)
```

```python
products.head()
```

```python
def great_fun(d):
    if 'great' in d:
        count = d['great']
    else:
        count = 0
    return count

products['great'] = products['word_count'].apply(great_fun)

products.head()
```

```python
def fantastic_fun(d):
    if 'fantastic' in d:
        count = d['fantastic']
    else:
        count = 0
    return count

products['fantastic'] = products['word_count'].apply(fantastic_fun)

products.head()
```

```python
def amazing_fun(d):
    if 'amazing' in d:
        count = d['amazing']
    else:
        count = 0
    return count

products['amazing'] = products['word_count'].apply(amazing_fun)

products.head()
```

```python
def love_fun(d):
    if 'love' in d:
        count = d['love']
    else:
        count = 0
    return count

products['love'] = products['word_count'].apply(love_fun)

products.head()
```

```python
def horrible_fun(d):
    if 'horrible' in d:
        count = d['horrible']
    else:
        count = 0
    return count

products['horrible'] = products['word_count'].apply(horrible_fun)

products.head()
```

```python
def hate_fun(d):
    if 'hate' in d:
        count = d['hate']
    else:
        count = 0
    return count

products['hate'] = products['word_count'].apply(hate_fun)

products.head()
```

## sentiment classifier

```python
train_data,test_data = products.random_split(0.8, seed =0)
```

```python
selected_words =['awesome', 'great', 'fantastic', 'amazing', 'bad', 'love', 'horrible', 'terrible','wow','awful', 'hate']
```

```python
products1 = products[products['rating'] != 3]
products1['sentiment'] = products1['rating'] >= 4
train_data,test_data = products1.random_split(0.8, seed =0)
selected_model = graphlab.logistic_classifier.create(train_data,target = 'sentiment', features= selected_words , validation_set=test_data)
```

```python
selected_model['coefficients'].sort(['value'],ascending = False)
```

```python
selected_model.evaluate(test_data, metric = 'roc_curve')
```

```python
selected_model.show(view = 'Evaluation')
```
